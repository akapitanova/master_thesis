```python
data_val = pd.read_csv('../data/val_data_stg7_norm.csv')
data_val.head()
```

```python
data_val['intensities'] = data_val['intensities'].apply(lambda x: eval(x) if isinstance(x, str) else x)
data_val['cond_vector'] = data_val['cond_vector'].apply(lambda x: eval(x) if isinstance(x, str) else x)
```

```python
i = 925
cond_vector = data_val.iloc[i]["cond_vector"]
vector = data_val.iloc[i]["intensities"]

plt.figure(figsize=(12, 6))

cond_str = ', '.join([f"{val:.2f}" for val in cond_vector])
plt.plot(wavelengths, vector, label=f"Predicted sample")
plt.xlabel("Wavelengths")
plt.ylabel("Intensity")
plt.title(f"Cond: [{cond_str}]")
plt.legend()
```
